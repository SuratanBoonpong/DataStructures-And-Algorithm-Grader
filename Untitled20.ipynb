```python
def convert_webp_to_jpg(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.webp'):
                webp_file_path = os.path.join(root, file)
                jpg_file_path = os.path.splitext(webp_file_path)[0] + '.jpg'

                img = Image.open(webp_file_path).convert('RGB')

                img.save(jpg_file_path, 'JPEG')

                print(f"Converted: {webp_file_path} to {jpg_file_path}")

convert_webp_to_jpg(directory)

print("Conversion completed.")
```

```python
root_directory = '/content/drive/MyDrive/image_dataset'

# Define the subdirectories for train, validation, and test
train_directory = os.path.join(root_directory, 'Train')
validation_directory = os.path.join(root_directory, 'Validate')
test_directory = os.path.join(root_directory, 'Test')

# Function to count the number of .jpg images in each class
def count_jpg_images_in_classes(directory):
    class_counts = {}
    for class_name in os.listdir(directory):
        class_path = os.path.join(directory, class_name)
        if os.path.isdir(class_path):
            jpg_images = [file for file in os.listdir(class_path) if file.endswith('.jpg')]
            num_images = len(jpg_images)
            class_counts[class_name] = num_images
    return class_counts

# Get the counts for train, validation, and test datasets
train_counts = count_jpg_images_in_classes(train_directory)
validation_counts = count_jpg_images_in_classes(validation_directory)
test_counts = count_jpg_images_in_classes(test_directory)

# Print the counts for each dataset
print("Train Dataset:")
for class_name, count in train_counts.items():
    print(f"Class '{class_name}': {count} .jpg images")

print("\nValidation Dataset:")
for class_name, count in validation_counts.items():
    print(f"Class '{class_name}': {count} .jpg images")

print("\nTest Dataset:")
for class_name, count in test_counts.items():
    print(f"Class '{class_name}': {count} .jpg images")
```

```python
base_model = MobileNet(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
x = Dense(1024, activation='relu')(x)
x = Dense(512, activation='relu')(x)

predictions = Dense(3, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

for layer in model.layers[:86]:
  layer.trainable=False #Freeze base model
for layer in model.layers[86:]:
  layer.trainable=True #Unfreeze new added denses

model.summary()
```

```python
def create_custom_model(X, Num_node_1,Num_node_2,Num_node_3):
    # Load base model
    base_model = MobileNet(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

    # Add new layers
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(Num_node_1, activation='relu')(x)
    x = Dense(Num_node_2, activation='relu')(x)
    x = Dense(Num_node_3, activation='relu')(x)

    predictions = Dense(3, activation='softmax')(x)

    model = Model(inputs=base_model.input, outputs=predictions)

    # Freeze the layers in the base model based on the input value X
    for layer in model.layers[:X]:
        layer.trainable = False  # Freeze base model
    for layer in model.layers[X:]:
        layer.trainable = True  # Unfreeze new added denses

    # Create Optimizer
    optimizer = Adam(learning_rate=0.0001,use_ema=True)

    # Compile the model
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

    model.summary()
    return model
```

```python
datagen = ImageDataGenerator(
    rotation_range=20,
    zoom_range=0.2,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
    preprocessing_function=preprocess_input,
    fill_mode="nearest"
)

batch_size = 64
seed_value = 42

train_generator = datagen.flow_from_directory(
    '/content/drive/MyDrive/image_dataset/Train',
    target_size=(224, 224),
    color_mode='rgb',
    batch_size=batch_size,
    class_mode='categorical',
    seed=seed_value,
    shuffle=True
)

val_generator = datagen.flow_from_directory(
    '/content/drive/MyDrive/image_dataset/Validate',
    target_size=(224, 224),
    color_mode='rgb',
    batch_size=batch_size,
    class_mode='categorical',
    seed=seed_value,
    shuffle=True
)

EP = 50
step_size_train=train_generator.n//train_generator.batch_size
step_size_val = val_generator.n//val_generator.batch_size

history = model.fit_generator(
    generator=train_generator,
    steps_per_epoch=step_size_train,
    validation_data=val_generator,
    validation_steps=step_size_val,
    epochs=EP,
    verbose=1
)
```

```python

N = len(history.history['accuracy'])
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(range(1, N + 1), history.history["accuracy"], label="Train Accuracy")
plt.plot(range(1, N + 1), history.history["val_accuracy"], label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training and Validation Accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(1, N + 1), history.history['loss'], label="Train Loss")
plt.plot(range(1, N + 1), history.history['val_loss'], label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Validation Loss")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
Test_generator = datagen.flow_from_directory(
    "/content/drive/MyDrive/image_dataset/Test",
    class_mode="categorical",
    target_size=(224, 224),
    color_mode="rgb",
    shuffle=False,
    batch_size=1
)
```

```python
y_true = Test_generator.classes
```

```python
preds = model.predict_generator(Test_generator)
```

```python
y_pred = np.argmax(preds,axis=1)
print(y_true)
print(y_pred)
```

```python
print("Shape of predictions:", preds.shape)

print("Predictions:")
print(preds)
```

```python
y_pred = np.argmax(preds, axis=1)
confusion = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:")
print(confusion)

report = classification_report(y_true, y_pred)
print("Classification Report:")
print(report)
```
